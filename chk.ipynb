```python
df2=pd.read_csv("Descriptor_calculated.csv")
df2
```

```python
df_final=pd.merge(df2,df,on='Smiles')
df_final
```

```python
# dropping useless columns from the dataset

df_final.drop(['Name','Test_ID','Smiles'],axis=1,inplace=True)
```

```python
# First 2 columns are also useless as all the entries are NaN

df_final.drop(['ABC','ABCGG'],axis=1,inplace=True)
```

```python
#since we are taking logSw as the target column hence Sweetness is also needed to be dropped

df_final.drop(['Sweetness'],axis=1,inplace=True)
df_final
```

```python
df_final.fillna(0,inplace=True)
```

```python
df_final
```

```python
X,y=df_final.iloc[:,:-1],df_final['logSw']
X
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# feature selection using f_regression

from sklearn.feature_selection import f_regression

f_stats, p_values = f_regression(X, y)
feature_indices = np.argsort(f_stats)[::-1]

num_selected_features =15
selected_feature_indices = feature_indices[:num_selected_features]
selected_features = X[:, selected_feature_indices]

selected_features
```

```python
X=pd.DataFrame(selected_features)
# X=pd.DataFrame(X)
```

```python
# X=pd.DataFrame(X)
```

```python
df=pd.concat([X,y],axis=1)
df
```

Model Implementation

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size=0.2, random_state=42)
```

```python
X_train,y_train = train.iloc[:, :-1],train['logSw']
X_test,y_test = test.iloc[:, :-1],test['logSw']
```

1. Linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print("Mean Absolute Error (MAE):", mae)
print("Root Mean Squared Error (RMSE):", (mse)**0.5)
```

```python
R = np.corrcoef(y_test, y_pred)[0, 1]
print("correlation_coefficient",R)
```
