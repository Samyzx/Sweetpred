```python
X,y=df_final_cont.iloc[:,:-1],df_final_cont.iloc[:,-1]
X
```

```python
from sklearn.preprocessing import StandardScaler

X_df = pd.DataFrame(X, columns=X.columns)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_df)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
X_scaled
```

```python
X_scaled=pd.DataFrame(X_scaled)
X_scaled.fillna(-9999999,inplace=True)

X_scaled=pd.DataFrame(X_scaled)
num_col=len(X_scaled.axes[1])-1
to_rem=[]
for i in range(num_col):
    lst=X_scaled[X_scaled.iloc[:,i].name].tolist()
    if((-9999999 in lst)):
        to_rem.append(X_scaled.iloc[:,i].name)

X_scaled.drop(to_rem, axis=1, inplace=True)
X_scaled
```

```python
# Applying covariance matrix to remove highly correlated features

# Step 1: Calculate the correlation matrix
corr_matrix = X_scaled.corr()

# Step 2: Identify highly correlated features
high_corr_features = set()
threshold = 0.7  # Adjust this threshold based on your analysis
for i in range(len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) > threshold:
            colname = corr_matrix.columns[i]
            high_corr_features.add(colname)
```

```python
len(high_corr_features)
```

```python
# Removing redundant features
X_scaled = X_scaled.drop(columns=high_corr_features)
X_scaled
```

```python
# Removing features not correlated with the target variable

#first, we will scale the target variable
scaler = StandardScaler()
y_scaled = scaler.fit_transform(pd.DataFrame(y))
y_scaled = pd.DataFrame(y_scaled, columns=['Label'])

#then, we will calculate the correlation between the target variable and the features
corr_with_target = X_scaled.copy()
corr_with_target['Label'] = y_scaled['Label']
correlation_matrix = corr_with_target.corr()
correlation_matrix = correlation_matrix['Label']
correlation_matrix = correlation_matrix.drop('Label')
```

```python
# Removing features not correlated with the target variable
threshold = 0.05
features_to_remove = correlation_matrix[abs(correlation_matrix) < threshold].index
X_scaled = X_scaled.drop(columns=features_to_remove)
X_scaled
```

Model Implementation

```python
# concatenating the target variable with the selected features
X_final = X_scaled_f_regressed
# nacid = X_scaled['nAcid']
# X_final['nAcid'] = nacid
X_final['Label'] = y

df = X_final
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size=0.2, random_state=42)
```

```python
X_train,y_train = train.iloc[:, :-1],train['Label']
X_test,y_test = test.iloc[:, :-1],test['Label']
```

1. Linear regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

model = LinearRegression()
model.fit(X_train, y_train)

y_pred_train = model.predict(X_train)

mae = mean_absolute_error(y_train, y_pred_train)
mse = mean_squared_error(y_train, y_pred_train)

print("ON TRAIN: ")
print("Mean Absolute Error (MAE):", mae)
print("Root Mean Squared Error (RMSE):", (mse)**0.5)
R = np.corrcoef(y_train, y_pred_train)[0, 1]
print("correlation_coefficient",R)


y_pred_test = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred_test)
mse = mean_squared_error(y_test, y_pred_test)

print("ON TEST: ")
print("Mean Absolute Error (MAE):", mae)
print("Root Mean Squared Error (RMSE):", (mse)**0.5)
R = np.corrcoef(y_test, y_pred_test)[0, 1]
print("correlation_coefficient",R)
```

2. Lasso Regression
